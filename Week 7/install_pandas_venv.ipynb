# Installing pandas in a Virtual Environment (venv)
This notebook provides step-by-step instructions to install pandas inside a virtual environment for both Windows and macOS/Linux.

## For Windows

Open a new terminal window and activate your VENV if it is not already on.




**Install pandas**

```python
pip install pandas
```

## For macOS/Linux

Open a new terminal window and activate your VENV if it is not already on. (see above section)




**Use this command to activate the virtual environment for mac**

```python
source venv/bin/activate
```

**Install pandas**

```python
pip install pandas
```
